```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'http://www.scrapethissite.com/pages/forms/'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup.prettify)
```

```python
soup.find('table')
```

```python
hockey_table = soup.find('table')
hockey_table
```

```python
hockey_table.find_all('th')
```

```python
titles = hockey_table.find_all('th')
titles
```

```python
hockey_table_titles = [title.text.strip() for title in titles]
hockey_table_titles
```

```python
df = pd.DataFrame(columns = hockey_table_titles)
df
```

```python
hockey_table.find_all('tr')
```

```python
hockey_table_row = hockey_table.find_all('tr')
```

```python
for row in hockey_table_row[1:]:
    raw_data = row.find_all('td')
    row_data = [data.text.strip() for data in raw_data]

    length = len(df)
    df.loc[length] = row_data
```
