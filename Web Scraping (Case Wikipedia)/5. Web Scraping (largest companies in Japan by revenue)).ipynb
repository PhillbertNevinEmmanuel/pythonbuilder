```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_Japanese_companies'
page = requests.get(url)
page_soup = BeautifulSoup(page.text, 'html')
```

```python
print(page_soup.prettify)
```

```python
page_soup.find_all('table')[0]
```

```python
table = page_soup.find_all('table')[0]
table
```

```python
table.find_all('th')
```

```python
table_header = table.find_all('th')
table_header
```

```python
table_header = [title.text.strip() for title in table_header]
table_header
```

```python
df = pd.DataFrame(columns = table_header)
df
```

```python
table.find_all('tr')
```

```python
raw_row_data = table.find_all('tr')
```

```python
for row in raw_row_data[1:]:
    raw_row = row.find_all('td')
    row_data = [data.text.strip() for data in raw_row]
    print(row_data)
    length = len(df)
    df.loc[length] = row_data
```
