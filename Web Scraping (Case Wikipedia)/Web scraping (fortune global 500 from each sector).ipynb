```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/Fortune_Global_500'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup.prettify)
```

```python
soup.find_all('table')[3]
```

```python
table = soup.find_all('table')[3]
```

```python
table
```

```python
table.find_all('th')
```

```python
header = table.find_all('th')
header
```

```python
table_header = [title.text.strip() for title in header]
table_header
```

```python
df = pd.DataFrame(columns = table_header)
df
```

```python
table.find_all('tr')
```

```python
raw_row_data = table.find_all('tr')
```

```python
for row in raw_row_data[1:]:
    raw_row = row.find_all('td')
    row_data = [data.text.strip() for data in raw_row]
    #print(row_data)
    row_insert = len(df)
    df.loc[row_insert] = row_data
```
